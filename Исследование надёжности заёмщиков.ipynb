```python
#–≤—Å–µ –ø—Ä–æ–ø—É—Å–∫–∏ –±—ã–ª–∏ –∏–∑–º–µ–Ω–µ–Ω—ã –Ω–∞ –Ω—É–ª–∏
#—Ç–µ–ø–µ—Ä—å —Ä–∞–∑–±–µ—Ä–µ–º—Å—è —Å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –≤ —Å—Ç–∞–∂–µ. –°–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ —ç—Ç–æ —Ç–µ—Ö–Ω–∏—á–µ—Å–∫–∞—è –æ—à–∏–±–∫–∞ –ø—Ä–∏
#–≤—ã–≥—Ä—É–∑–∫–µ –±–∞–∑—ã, –¥—Ä—É–≥–æ–≥–æ –æ–±—ä—è—Å–Ω–µ–Ω–∏—è –Ω–µ –≤–∏–∂—É, –Ω–æ –∏—Å–ø—Ä–∞–≤–ª—è—Ç—å –Ω–∞–¥–æ. –ò—Å–ø—Ä–∞–≤–∏–º –Ω–∞ –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.
bk['experience'] = bk['experience'].abs()
bk['experience'].describe()
```

```python
#—Ç–µ–ø–µ—Ä—å —Å–º—É—â–∞–µ—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –Ω–∞ —Ñ–æ–Ω–µ —Å—Ä–µ–¥–Ω–µ–≥–æ. –î–ª—è –≤–∏–∑—É–∞–ª—å–Ω–æ–≥–æ —É–¥–æ–±—Å—Ç–≤–∞ –ø–µ—Ä–µ–≤–µ–¥–µ–º –¥–Ω–∏
#—Å—Ç–∞–∂–∞ –≤ –≥–æ–¥—ã –∏ –ø–æ–º–µ—Å—Ç–∏–º –∏—Ö –≤ –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü

bk['years_experience'] = bk['experience'] / 365
```

```python
#—Å—É–¥—è –ø–æ –≤—ã–≤–æ–¥—É –≤ 932 –≥–æ–¥–∞ –≤ —Å—Ç–∞–∂–µ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Ç–∞–∫–∂–µ –∑–∞–∫—Ä–∞–ª–∏—Å—å –æ—à–∏–±–∫–∏. –ü–æ—Å–º–æ—Ç—Ä–∏–º —Å–∫–æ–ª—å–∫–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —É –Ω–∞—Å
#–≤—ã—à–ª–∏ –∑–∞ —Ä–∞–º–∫–∏ –≤–æ–∑–º–æ–∂–Ω–æ–≥–æ, –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞—è —á—Ç–æ –∫–ª–∏–µ–Ω—Ç –Ω–∞—á–∞–ª —Ä–∞–±–æ—Ç–∞—Ç—å —Å 16 –ª–µ—Ç –∏ –≤—ã—à–µ–ª –Ω–∞ –ø–µ–Ω—Å–∏—é –≤ 60.

mistakes = bk[bk['years_experience'] > 44].count()[0]
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å–æ —Å—Ç–∞–∂–µ–º –±–æ–ª–µ–µ 44 –ª–µ—Ç: ", mistakes)
share_mistakes = mistakes / bk['years_experience'].count()
print("–î–æ–ª—è –æ—à–∏–±–æ–∫ –≤ —Å–ø–∏—Å–∫–µ: " '{:.2%}'.format(share_mistakes))
```

- –ø—Ä–æ–≤–µ—Ä–∫–∞ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è –æ–±—ä–µ–∫—Ç–∞ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –ø—É—Ç–∏;
- –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ –≤ –±–∞–π—Ç–∞—Ö;

```python
#–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –º–Ω–æ–≥–æ –¥–ª—è —Ç–µ—Ö–Ω–∏—á–µ—Å–∫–æ–π –æ—à–∏–±–∫–∏, –ø–æ–∫–∞ –Ω–µ–ø–æ–Ω—è—Ç–Ω–æ —á—Ç–æ —Å —ç—Ç–∏–º –¥–µ–ª–∞—Ç—å, –≤–µ—Ä–Ω–µ–º—Å—è –∫ —ç—Ç–∏–º –¥–∞–Ω–Ω—ã–º
#–ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –∏ –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏–∏

#—Ç–µ–ø–µ—Ä—å –ø—Ä–∏–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ —É—Ä–æ–≤–Ω–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –∫ –µ–¥–∏–Ω–æ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
bk['education'] = bk['education'].str.lower()
bk['education'].unique()
```

```python
#–∫–∞–∫ –∏ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å, –ø–æ–ª—É—á–∏–ª–æ—Å—å –ø—è—Ç—å —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —É—Ä–æ–≤–Ω–µ–π.

#–¥–∞–ª–µ–µ –±—É–¥–µ–º —Ä–∞–∑–∏—Ä–∞—Ç—å—Å—è —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –¥–µ—Ç–µ–π. –ü–æ—Å–º–æ—Ç—Ä–∏–º —Å–Ω–∞—á–∞–ª–∞ –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç—Å–≤–æ –Ω–µ–ø–æ–Ω—è—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π.
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å  0: ", bk[bk['children'] == 0].count()[0])
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å  -1: ", bk[bk['children'] == -1].count()[0])
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å  20: ", bk[bk['children'] == 20].count()[0])
```

```python
#—Å–Ω–æ–≤–∞ –Ω–∞–±–ª—é–¥–∞—é—Ç—Å—è –æ—à–∏–±–∫–∏, –¥–∞–Ω–Ω—ã–µ –ø–æ 47 –∫–ª–∏–µ–Ω—Ç–∞–º –æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –¥–µ—Ç–µ–π –∑–∞–º–µ–Ω–∏–º –Ω–∞ 0,
#–∞–Ω–∞–ª–∏–∑ –æ—Ç —ç—Ç–æ–≥–æ –Ω–µ –ø–æ—Å—Ç—Ä–∞–¥–∞–µ—Ç, –∞ –ø–æ 76 –∫–ª–∏–µ–Ω—Ç–∞–º –∏–∑–º–µ–Ω–∏–º –∑–Ω—á–µ–Ω–∏—è –Ω–∞ 2 - —Å–æ–≥–ª–∞—Å–Ω–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏,
# –∏–∑ –æ—Ç–∫—Ä—ã—Ç—ã–∑—Ö –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤ –∏–Ω—Ñ–æ—Ä–∞—Ü–∏–∏ –≤ –∏–Ω—Ç—Ä–µ–Ω–µ—Ç–µ, –≤ –†–§ –Ω–µ—Ç —Å—Ç–æ–ª—å–∫–æ —Å–µ–º–µ–π —Å 20 –¥–µ—Ç—å–º–∏. 

bk['children'] = bk['children'].replace(-1, 1)
bk['children'] = bk['children'].replace(20, 2)
bk['children'].unique()
```

–û—á–µ–Ω—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ –∑–∞–º–µ—á–∞–Ω–∏–µ –∏ –æ—Ç–ª–∏—á–Ω–æ –ø—Ä–∏–º–µ–Ω–µ–Ω –º–µ—Ç–æ–¥ `replace` –¥–ª—è –∑–∞–º–µ–Ω—ã

```python
#–¥–∞–ª–µ–µ –Ω–∞ –æ—á–µ—Ä–µ–¥–∏ –≤–æ–∑—Ä–∞—Å—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å –Ω—É–ª–µ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ

#—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–∏–º –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç—É —É—Å—Ç–æ—è–≤—à–∏–º—Å—è –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º - –¥–æ 18 –ª–µ—Ç, –æ—Ç 18 –¥–æ 65,
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ –≤–æ–∑–≤—Ä—Å—Ç–µ 0: ", bk[bk['age'] == 0].count()[0])
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ –≤–æ–∑–≤—Ä—Å—Ç–µ –¥–æ 18: ", bk[bk['age'] < 18].count()[0])
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ –≤–æ–∑–≤—Ä—Å—Ç–µ –æ—Ç 18: ", bk[bk['age'] >= 19].count()[0])
```

```python
#–∫–∞–∫ –≤–∏–¥–Ω–æ –∏–∑ —Ä–∞—Å—á–µ—Ç–æ–≤, "–Ω—É–ª–µ–≤—ã–µ" –∫–ª–∏–µ–Ω—Ç—ã –∏ —Ç–æ–ª—å–∫–æ –¥–æ—Å—Ç–∏–≥—à–∏–µ —Å–æ–≤–µ—Ä—à–µ–Ω–Ω–æ–ª–µ—Ç–∏–µ - —ç—Ç–æ –æ–¥–∏–Ω –∏ —Ç–µ–∂–µ –ª—é–¥–∏.
#–ø–µ—Ä–µ–≤–µ–¥–µ–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ –≤ —Ç–∏–ø "int" –∏ —Å—Ä–∞–∑—É –ø—Ä–æ–≤–µ—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω—É–ª–µ–π
bk['age'] = bk.groupby('profession')['age'].transform(lambda x: x.replace(0, int(x.mean())))
print('–†–µ–∑—É–ª—å—Ç–∞—Ç: ', bk[bk['age'] == 0].count()[0])
```

```python
# –ø—Ä–æ–≤–µ–¥–µ–º –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏—é,
# –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –∑–∞–º–µ–Ω–∏–≤ –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –ª–µ–º–º –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –Ω–∞ –≥–ª–∞–≤–Ω–æ–µ –∫–ª—é—á–µ–≤–æ–µ —Å–ª–æ–≤–æ –∏–∑ —Å–ø–∏—Å–∫–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–π

def lemmatize(text):
    lemma = m.lemmatize(text)
    for word in categories:
        if word in lemma:
            lemma = word
    return lemma

bk['purpose_group'] = bk['purpose'].apply(lemmatize)        
bk.head(5)
```

```python
#—Ä–∞–Ω–µ–µ –ø–æ—Å–ª–µ –ª–µ–º–∞—Ç–∏–∑–∞—Ü–∏–∏ —É–∂–µ –±—ã–ª–∞ –≤—ã–ø–æ–ª–Ω–µ–Ω–∞ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Ü–µ–ª–∏ –∫—Ä–µ–¥–∏—Ç–∞, –∏ –¥–æ–±–∞–≤–ª–µ–Ω –≤ —Ç–∞–±–ª–∏—Ü—É —Å—Ç–æ–ª–±–µ—Ü 
#—Å –Ω–æ–≤—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º. –î–∞–ª–µ–µ —Å–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π, —Å–µ–º–µ–π–Ω–æ–º—É –ø–æ–ª–æ–∂–µ–Ω–∏—é –∏ –¥–æ—Ö–æ–¥—É

#–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π
children_group = pd.DataFrame()
children_group['count_children'] = bk.groupby('children')['debt'].count()
children_group['share_children'] = children_group['count_children'] * 100 / bk['children'].count()
children_group['share_children'] = round(children_group['share_children'], 2).astype(str) + '%'
children_group
```

```python
#—Ç–∞–∫ –∫–∞–∫ –ø–æ –∑–∞–¥–∞–Ω–∏—é –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —É–∑–Ω–∞—Ç—å –∫–∞–∫ –≤–ª–∏—è–µ—Ç –Ω–∞–ª–∏—á–∏–µ –¥–µ—Ç–µ–π –Ω–∞ –≤–æ–∑–≤—Ä–∞—Ç –∫—Ä–µ–¥–∏—Ç–∞ –≤ —Å—Ä–æ–∫, 
#–≤—ã–¥–µ–ª–∏–º –¥–≤–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –≥—Ä—É–ø–ø—ã –∏ –¥–æ–±–∞–≤–∏–º –∏—Ö –≤ —Ç–∞–±–ª–∏—Ü—É –Ω–æ–≤—ã–º —Å—Ç–æ–ª–±—Ü–æ–º
def children_count(children):
    if children == 0:
            return '–ë–µ–∑ –¥–µ—Ç–µ–π'
    return '–° –¥–µ—Ç—å–º–∏'
```

```python
bk['children_count'] = bk['children'].apply(children_count)
```

```python
#—Å–¥–µ–ª–∞–µ–º —Å–≤–æ–¥–Ω—É—é –ø–æ–≤—Ç–æ—Ä–Ω–æ
children_count_group = pd.DataFrame()
children_count_group['count_children'] = bk.groupby('children_count')['debt'].count()
children_count_group['share_children'] = children_count_group['count_children'] * 100 / bk['children_count'].count()
children_count_group['share_children'] = round(children_count_group['share_children'], 2).astype(str) + '%'
children_count_group
```

**- –ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –ø–æ —Å–µ–º–µ–π–Ω–æ–º—É –ø–æ–ª–æ–∂–µ–Ω–∏—é**

```python
#–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Å–µ–º–µ–π–Ω–æ–º—É –ø–æ–ª–æ–∂–µ–Ω–∏—é

family_group = pd.DataFrame()
family_group['count_family'] = bk.groupby('family_status')['debt'].count()
family_group['share_family'] = family_group['count_family'] * 100 / bk['family_status'].count()
family_group['share_family'] = round(family_group['share_family'], 2).astype(str) + '%'
family_group = family_group.sort_values(by ='count_family', ascending = 0)
family_group
```

```python
#–≤ –∫–∞—á–µ—Å—Ç–≤–µ –≥—Ä–∞–Ω–∏—Ü –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –¥–æ—Ö–æ–¥–æ–≤ –∏—Å–ø–æ–ª—å–∑—É–µ–º –∫–≤–∞–Ω—Ç–∏–ª–∏ (25%, 50% –∏ 75%).

bk['income'].quantile([0.25,0.5,0.75])
```

```python
#—Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ –æ–ø—Ä–µ–¥–µ–ª–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∏ –¥–æ–±–∞–≤–∏–º –∏—Ö –Ω–æ–≤—ã–º —Å—Ç–æ–ª–±—Ü–æ–º –≤ —Ç–∞–±–ª–∏—Ü—É

def income_status(income):
    if income <= 89088.50:
            return '–ù–∏–∑–∫–∏–π —É—Ä–æ–≤–µ–Ω—å'
    if income <= 135781.0:
            return '–°—Ä–µ–¥–Ω–∏–π —É—Ä–æ–≤–µ–Ω—å'
    if income < 195813.25:
            return '–í—ã—Å–æ–∫–∏–π —É—Ä–æ–≤–µ–Ω—å'
    return '–°–≤–µ—Ä—Ö–≤—ã—Å–æ–∫–∏–π —É—Ä–æ–≤–µ–Ω—å'
```

```python
bk['income_status'] = bk['income'].apply(income_status)
```

```python
#–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —É—Ä–æ–≤–Ω—é –¥–æ—Ö–æ–¥–∞
income_status_group = pd.DataFrame()
income_status_group['count_status'] = bk.groupby('income_status')['debt'].count()
income_status_group['share_income'] = income_status_group['count_status'] * 100 / bk['income_status'].count()
income_status_group['share_income'] = round(income_status_group['share_income'], 3).astype(str) + '%'
income_status_group
```
